# Google Search Sessions and Orders Monthly Trend

```python
%%sql
WITH data_trend AS (SELECT YEAR(ws.created_at) yearno, MONTH(ws.created_at) AS monthno, ws.website_session_id, o.order_id
FROM website_sessions AS ws
LEFT JOIN orders AS o
USING (website_session_id)
WHERE ws.utm_source = 'gsearch' AND
ws.created_at < '2012-11-27')

SELECT dt.yearno AS 'year', dt.monthno AS 'month', COUNT(DISTINCT dt.website_session_id) AS sessions, COUNT(DISTINCT dt.order_id) AS orders, COUNT(DISTINCT dt.order_id) / COUNT(DISTINCT dt.website_session_id) AS conv_rate
FROM data_trend AS dt
GROUP BY dt.yearno, dt.monthno
```

```python
%%sql
WITH data_trend AS (SELECT YEAR(ws.created_at) yearno, MONTH(ws.created_at) AS monthno, ws.website_session_id, ws.utm_campaign, o.order_id
FROM website_sessions AS ws
LEFT JOIN orders AS o
USING (website_session_id)
WHERE ws.utm_source = 'gsearch' AND
ws.created_at < '2012-11-27' AND
ws.utm_campaign IN ('brand', 'nonbrand'))

SELECT dt.yearno AS 'year', dt.monthno AS 'month', dt.utm_campaign, COUNT(DISTINCT dt.website_session_id) AS sessions, COUNT(DISTINCT dt.order_id) AS orders,COUNT(DISTINCT dt.order_id)/ COUNT(DISTINCT dt.website_session_id) AS conv_rate
FROM data_trend AS dt
GROUP BY dt.yearno, dt.monthno, dt.utm_campaign
```

# Sessions and Orders Monthly Trend By Source

```python
%%sql
WITH website_sessions AS (
    SELECT 
    YEAR(ws.created_at) yearno,
    MONTH(ws.created_at) AS monthno,
    CASE WHEN ws.utm_source = 'gsearch' THEN 1 ELSE 0 END AS gsearch_session,
    CASE WHEN ws.utm_source = 'bsearch' THEN 1 ELSE 0 END AS bsearch_session,
    CASE WHEN ws.utm_source IS NULL AND ws.http_referer IS NOT NULL THEN 1 ELSE 0 END AS organic_session,
    CASE WHEN ws.utm_source IS NULL AND ws.http_referer IS NULL THEN 1 ELSE 0 END AS typein_session,
    CASE WHEN ws.utm_source = 'gsearch' AND o.order_id IS NOT NULL THEN 1 ELSE 0 END AS gsearch_order,
    CASE WHEN ws.utm_source = 'bsearch' AND o.order_id IS NOT NULL THEN 1 ELSE 0 END AS bsearch_order,
    CASE WHEN ws.utm_source IS NULL AND ws.http_referer IS NOT NULL AND o.order_id IS NOT NULL THEN 1 ELSE 0 END AS organic_order,
    CASE WHEN ws.utm_source IS NULL AND ws.http_referer IS NULL AND o.order_id IS NOT NULL THEN 1 ELSE 0 END AS typein_order
    
FROM website_sessions AS ws
LEFT JOIN orders AS o
USING (website_session_id)
WHERE ws.created_at < '2012-11-27'
)

SELECT yearno, monthno, SUM(gsearch_session) AS gsearch_sessions, SUM(bsearch_session) AS bsearch_sessions, SUM(organic_session) AS organic_sessions, SUM(typein_session) AS typein_sessions, SUM(gsearch_order) AS gsearch_orders, SUM(bsearch_order) AS bsearch_orders, SUM(organic_order) AS organic_orders, SUM(typein_order) AS typein_orders
FROM website_sessions
GROUP BY yearno, monthno
```

# Lander-1 Additional Orders Estimate

```python
%%sql
WITH lander_convrt AS (SELECT
    wpv.pageview_url,
    COUNT(DISTINCT ws.website_session_id) AS sessions,
    COUNT(DISTINCT o.order_id) AS orders,
    COUNT(DISTINCT o.order_id) / COUNT(DISTINCT ws.website_session_id) AS conv_rate,
    SUM(o.price_usd) AS revenue
FROM website_sessions AS ws
LEFT JOIN orders AS o
    USING (website_session_id)
LEFT JOIN website_pageviews AS wpv
    USING (website_session_id)

WHERE
    ws.utm_source = 'gsearch' AND
    ws.utm_campaign = 'nonbrand' AND
    ws.created_at BETWEEN (SELECT MIN(created_at) FROM website_pageviews WHERE pageview_url = '/lander-1') AND '2012-07-28' AND
    wpv.pageview_url IN ('/home', '/lander-1')
GROUP BY wpv.pageview_url)

SELECT
    ROUND(COUNT(DISTINCT website_session_id) * (SELECT (SELECT conv_rate FROM lander_convrt WHERE pageview_url = '/lander-1') - (SELECT conv_rate FROM lander_convrt WHERE pageview_url = '/home')),0) AS extra_orders_since_lander1
FROM website_sessions AS ws
LEFT JOIN website_pageviews AS wpv
USING(website_session_id)
WHERE
    ws.utm_source = 'gsearch' AND
    ws.utm_campaign = 'nonbrand' AND
    ws.created_at < '2012-11-27' AND
    wpv.pageview_url IN ('/home', '/lander-1') AND
    ws.created_at BETWEEN
        (SELECT MAX(wpv.created_at)
            FROM website_sessions AS ws
            LEFT JOIN orders AS o
            USING (website_session_id)
            LEFT JOIN website_pageviews AS wpv
            USING (website_session_id)
            WHERE 
                ws.utm_source = 'gsearch' AND
                ws.utm_campaign = 'nonbrand' AND
                wpv.pageview_url = '/home'
            GROUP BY wpv.pageview_url) AND
        '2012-11-27'
```

# Google Search Sessions Lander Conversion Funnel Comparison

## Creating Hits Table

```python
%%sql
CREATE TEMPORARY TABLE page_hits
WITH site_flags AS(
SELECT ws.website_session_id,
CASE WHEN wpv.pageview_url = '/products' THEN 1 ELSE NULL END AS products_flag,
CASE WHEN wpv.pageview_url = '/the-original-mr-fuzzy' THEN 1 ELSE NULL END AS fuzzy_flag,
CASE WHEN wpv.pageview_url = '/cart' THEN 1 ELSE NULL END AS cart_flag,
CASE WHEN wpv.pageview_url = '/shipping' THEN 1 ELSE NULL END AS shipping_flag,
CASE WHEN wpv.pageview_url = '/billing' THEN 1 ELSE NULL END AS billing_flag,
CASE WHEN wpv.pageview_url = '/thank-you-for-your-order' THEN 1 ELSE NULL END AS thank_flag

FROM website_sessions AS ws
LEFT JOIN website_pageviews AS wpv
USING(website_session_id)
WHERE ws.created_at BETWEEN '2012-06-19' AND '2012-07-28'),





session_pages AS (
SELECT
    site_flags.website_session_id,
    MAX(products_flag) AS acessed_products,
    MAX(fuzzy_flag) AS acessed_fuzzy,
    MAX(cart_flag) AS acessed_cart,
    MAX(shipping_flag) AS acessed_shipping,
    MAX(billing_flag) AS acessed_billing,
    MAX(thank_flag) AS acessed_thank   
FROM site_flags
GROUP BY site_flags.website_session_id)





SELECT 
    wpv.pageview_url,
    COUNT(DISTINCT sp.website_session_id) AS sessions,
    SUM(sp.acessed_products) AS products_hits,
    SUM(sp.acessed_fuzzy) AS fuzzy_hits,
    SUM(sp.acessed_cart) AS cart_hits,
    SUM(sp.acessed_shipping) AS shipping_hits,
    SUM(sp.acessed_billing) AS billing_hits,
    SUM(sp.acessed_thank) AS thankyou_hits             

FROM website_sessions AS ws

LEFT JOIN website_pageviews AS wpv
    USING(website_session_id)

LEFT JOIN session_pages AS sp
    USING (website_session_id)

WHERE 
    ws.created_at BETWEEN '2012-06-19' AND '2012-07-28'AND
    ws.utm_source = 'gsearch' AND
    ws.utm_campaign = 'nonbrand' AND
    wpv.pageview_url IN ('/lander-1', '/home')
GROUP BY wpv.pageview_url
```

### Hits table

```python
%%sql
SELECT * FROM page_hits
```

### Funnel conversion rate

```python
%%sql
SELECT
    ph.pageview_url,
    ph.products_hits / ph.sessions AS lander_conv_rt,
    ph.fuzzy_hits / ph.products_hits AS products_pg_conv_rt,
    ph.cart_hits / ph.fuzzy_hits AS fuzzy_pg_conv_rt,
    ph.shipping_hits / ph.cart_hits AS cart_pg_conv_rt,
    ph.billing_hits / ph.shipping_hits AS shipping_pg_conv_rt,
    ph.thankyou_hits / ph.billing_hits AS billing_pg_conv_rt
FROM page_hits AS ph
```

# Billing Test Impact Analysis

```python
%%sql
WITH billing_test AS
(SELECT 
    wpv.pageview_url,
    COUNT(DISTINCT ws.website_session_id) AS sessions,
    COUNT(DISTINCT o.order_id) AS orders,
    ROUND(SUM(o.price_usd) / COUNT(DISTINCT ws.website_session_id), 2) AS revenue_per_billing_page_seen

FROM website_sessions AS ws
LEFT JOIN website_pageviews AS wpv
USING(website_session_id)
LEFT JOIN orders AS o
USING(website_session_id)

WHERE 
    ws.created_at BETWEEN '2012-09-10' AND '2012-11-10'AND
    wpv.pageview_url IN ('/billing', '/billing-2')

GROUP BY wpv.pageview_url)

SELECT *
FROM billing_test
LEFT JOIN(
SELECT wpv.pageview_url, COUNT(ws.website_session_id) AS past_month_sessions
FROM website_sessions AS ws
LEFT JOIN website_pageviews AS wpv
USING(website_session_id)
WHERE 
    ws.created_at BETWEEN '2012-10-27' AND '2012-11-27'AND
    wpv.pageview_url IN ('/billing', '/billing-2')
GROUP BY wpv.pageview_url
) AS past_month_billing_test
USING(pageview_url)
UNION                         -- Manual calculation input, just for fun.
(
SELECT 'TOTAL', '1311', COUNT(DISTINCT o.order_id) AS orders, '$ LIFT: 8.51', COUNT(ws.website_session_id) AS sessions
FROM website_sessions AS ws
LEFT JOIN website_pageviews AS wpv
USING(website_session_id)
LEFT JOIN orders AS o
USING(website_session_id)
WHERE 
    ws.created_at BETWEEN '2012-10-27' AND '2012-11-27'AND
    wpv.pageview_url IN ('/billing', '/billing-2')
)
```
